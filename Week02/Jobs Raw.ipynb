Seprate Salary and location column

```python
import re
import pandas as pd

def extract_location_and_salary(location_str):
    if pd.isna(location_str):
        return None, None

    location_str = str(location_str).strip()

    # Define a regex pattern to identify the start of salary information
    # This pattern looks for:
    # 1. A currency symbol (£, $, €, kr, dkk)
    # 2. A digit followed by 'K' or 'M' (e.g., '30K', '40M')
    # 3. Common salary-related phrases (e.g., '(Employer provided)', 'per annum')
    salary_start_pattern = re.compile(
        r'(\b£|\b\$|\b€|\bkr|\bdkk|\b\d+[KkMm]\b|\(Employer provided\)|\bper\s*annum\b|\bp\.a\.\b|\byearly\b|\bpa\b|\bannual\b)',
        re.IGNORECASE
    )

    match = salary_start_pattern.search(location_str)

    if match:
        # Everything from the start of the match till the end is potentially salary
        salary_part = location_str[match.start():].strip()
        # Everything before the start of the match is location
        location_part = location_str[:match.start()].strip()

        # Clean up trailing commas or whitespace from location_part if any
        if location_part.endswith(','):
            location_part = location_part[:-1].strip()

        return location_part if location_part else None, salary_part if salary_part else None
    else:
        # If no salary pattern is found, the whole string is considered location
        return location_str if location_str else None, None

# Apply the function to the 'location' column and overwrite 'location' and 'salary' columns
df[['location', 'salary']] = df['location'].apply(lambda x: pd.Series(extract_location_and_salary(x)))

# Display the updated 'location' and 'salary' columns
display(df[['location', 'salary']].head(20))
```

Job Type Filling

```python
def extract_job_type(description):
    if pd.isna(description):
        return None

    description = str(description).lower()

    # Define keywords for common job types
    job_type_keywords = {
        'full-time': ['full-time', 'full time'],
        'part-time': ['part-time', 'part time'],
        'contract': ['contract', 'contractual'],
        'temporary': ['temporary', 'temp'],
        'internship': ['internship', 'intern'],
        'apprenticeship': ['apprenticeship', 'apprentice'],
        'permanent': ['permanent'],
        'freelance': ['freelance', 'freelancer']
    }

    found_types = []
    for job_type, keywords in job_type_keywords.items():
        for keyword in keywords:
            if keyword in description:
                found_types.append(job_type)
                # If multiple keywords for the same type, only add once
                break

    # Prioritize if multiple types are found, or return the first unique one
    # For simplicity, returning the first found type. Can be refined for complex cases.
    if found_types:
        return ', '.join(sorted(list(set(found_types)))) # Return unique and sorted types
    else:
        return 'unspecified'

# Apply the function to the 'description' column to fill the 'job_type' column
df['job_type'] = df['description'].apply(extract_job_type)

# Display the value counts of the newly populated 'job_type' column
display(df['job_type'].value_counts(dropna=False))
```

Education Level

```python
def extract_education_level(description):
    if pd.isna(description):
        return None

    description = str(description).lower()

    education_keywords = {
        'bachelor': ['bachelor', 'ba', 'bs', 'b.a.', 'b.s.', 'undergraduate'],
        'master': ['master', 'ma', 'ms', 'mba', 'm.a.', 'm.s.', 'graduate'],
        'phd': ['phd', 'doctorate', 'dphil', 'doctor'],
        'associate': ['associate', 'aa', 'as', 'a.a.', 'a.s.'],
        'high school': ['high school', 'ged', 'diploma'],
        'vocational': ['vocational', 'certificate', 'diploma'], # Diploma can be general, but often implies vocational
        'degree': ['degree'] # General degree mention if specific level not found
    }

    found_levels = []
    for level, keywords in education_keywords.items():
        for keyword in keywords:
            if keyword in description:
                found_levels.append(level)
                break

    if found_levels:
        # Prioritize more specific levels if multiple are found
        # For simplicity, returning unique and sorted levels here.
        # A more complex prioritization logic could be implemented if needed.
        return ', '.join(sorted(list(set(found_levels))))
    else:
        return 'unspecified'

# Apply the function to the 'description' column to fill the 'education_level' column
df['education_level'] = df['description'].apply(extract_education_level)

# Display the value counts of the newly populated 'education_level' column
display(df['education_level'].value_counts(dropna=False))
```

Skills

```python
import re
import pandas as pd

SKILL_TRIGGER_PATTERNS = [
    r"skills include[:\-]?\s*(.*?)(?:\.|\n|$)",
    r"experience with[:\-]?\s*(.*?)(?:\.|\n|$)",
    r"knowledge of[:\-]?\s*(.*?)(?:\.|\n|$)",
    r"proficient in[:\-]?\s*(.*?)(?:\.|\n|$)",
    r"familiar with[:\-]?\s*(.*?)(?:\.|\n|$)",
    r"working knowledge of[:\-]?\s*(.*?)(?:\.|\n|$)"
]

SKILL_SPLIT_REGEX = re.compile(
    r",|;|/|\band\b|\bor\b", re.IGNORECASE
)

# Explicit salary / money blocker
SALARY_REGEX = re.compile(
    r"£|\$|€|\d{2,}[,.]\d+|\d+\s*(per|/)\s*(year|hour|annum)",
    re.IGNORECASE
)

def extract_skills_generic(text):
    if pd.isna(text):
        return None

    skills = set()

    for pattern in SKILL_TRIGGER_PATTERNS:
        for match in re.findall(pattern, text, flags=re.IGNORECASE | re.DOTALL):
            parts = SKILL_SPLIT_REGEX.split(match)

            for part in parts:
                skill = part.strip().lower()

                # HARD FILTERS
                if not (2 <= len(skill) <= 40):
                    continue
                if SALARY_REGEX.search(skill):
                    continue
                if skill.count('%') > 0:
                    continue
                if sum(c.isdigit() for c in skill) > 2:
                    continue
                if any(x in skill for x in [
                    "we offer", "benefits", "holiday", "salary", "package"
                ]):
                    continue

                skills.add(skill)

    return ", ".join(sorted(skills)) if skills else None


# Apply
df['skills'] = df['description'].apply(extract_skills_generic)
```

Experience Level

```python
import re
import pandas as pd

EXPERIENCE_PATTERNS = {
    "intern": [
        r"\bintern(ship)?\b",
        r"\bstudent\b"
    ],
    "entry": [
        r"\bgraduate\b",
        r"\bjunior\b",
        r"\bentry[- ]level\b",
        r"\bno experience\b",
        r"\btraining provided\b"
    ],
    "mid": [
        r"\bintermediate\b",
        r"\bmid[- ]level\b",
        r"\b2\+?\s*years\b",
        r"\b3\+?\s*years\b"
    ],
    "senior": [
        r"\bsenior\b",
        r"\blead\b",
        r"\bprincipal\b",
        r"\b5\+?\s*years\b",
        r"\b6\+?\s*years\b"
    ],
    "management": [
        r"\bmanager\b",
        r"\bhead of\b",
        r"\bdirector\b"
    ],
    "executive": [
        r"\bvp\b",
        r"\bvice president\b",
        r"\bchief\b",
        r"\bexecutive\b"
    ]
}

def extract_experience_level(text):
    if pd.isna(text):
        return None

    text = text.lower()

    detected_levels = []

    for level, patterns in EXPERIENCE_PATTERNS.items():
        for pattern in patterns:
            if re.search(pattern, text):
                detected_levels.append(level)
                break

    # Priority order (most senior wins)
    PRIORITY = [
        "executive",
        "management",
        "senior",
        "mid",
        "entry",
        "intern"
    ]

    for level in PRIORITY:
        if level in detected_levels:
            return level

    return None


# Apply to dataframe
df['experience_level'] = df['description'].apply(extract_experience_level)
```
